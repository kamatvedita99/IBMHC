```python
import twint
import nest_asyncio
```

```python
nest_asyncio.apply()
c = twint.Config()
c.Search = "lockdown" 
c.Since = "2020-04-19"
c.Until = "2020-05-04"
c.Pandas = True
c.Geo = "28.5934,77.2223,2000km"# Run
twint.run.Search(c)
```

```python
df = twint.storage.panda.Tweets_df
df
```
